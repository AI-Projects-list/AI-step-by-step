# 1. What is Model Context Protocol (MCP)?

## ğŸ“– What is MCP?

**Model Context Protocol (MCP)** is an open protocol that standardizes how AI applications provide context to Large Language Models (LLMs). It enables seamless integration between AI models and various data sources, tools, and services.

**Core Concept:**
```
AI Application (Client) â†â†’ MCP Protocol â†â†’ Context Providers (Servers)

Example:
Claude Desktop â†â†’ MCP â†â†’ [GitHub, Filesystem, Database, Slack]
```

**Key Components:**
```
1. MCP Client
   - AI application (Claude, custom chatbot)
   - Requests resources and invokes tools

2. MCP Server
   - Context provider (filesystem, database, API)
   - Exposes resources, tools, prompts

3. Protocol Messages
   - JSON-RPC 2.0 format
   - Request/Response pattern

4. Transport Layer
   - stdio (local processes)
   - HTTP/SSE (remote servers)
   - WebSocket (bidirectional)
```

**What MCP Provides:**
- **Resources**: Data sources (files, databases, APIs)
- **Tools**: Actions the LLM can perform (search, create, update)
- **Prompts**: Pre-defined prompt templates with context
- **Sampling**: Request LLM completions from the client

## ğŸ¯ Why Use MCP?

### **Advantages:**
1. **Standardization** - One protocol for all context providers
2. **Interoperability** - Any client works with any server
3. **Security** - Sandboxed, controlled access to resources
4. **Modularity** - Add/remove context sources easily
5. **Scalability** - Connect to unlimited data sources
6. **Flexibility** - Support local and remote servers
7. **Open Source** - Community-driven, extensible

### **Problems MCP Solves:**
1. **Fragmented Integrations** - Each tool needs custom integration
2. **Context Limitations** - LLMs have limited context windows
3. **Real-Time Data** - Models can't access live data
4. **Tool Complexity** - Hard to give LLMs tool access safely
5. **Vendor Lock-In** - Proprietary protocols limit choices

## â±ï¸ When to Use MCP

### âœ… **Use MCP When:**

**1. Building AI Assistants**
- Example: Customer support bot needing access to knowledge base
- Why: Cleanly separate AI logic from data sources
- Benefit: Swap data sources without changing AI code

**2. Integrating Multiple Data Sources**
- Example: AI needs GitHub + Jira + Slack + Database
- Why: One protocol handles all integrations
- Benefit: Add new sources in minutes, not days

**3. Grounding LLMs with Real-Time Data**
- Example: AI answering questions about live inventory
- Why: Prevent hallucinations with current facts
- Benefit: Accurate, up-to-date responses

**4. Enabling Tool Use (Function Calling)**
- Example: AI that can create GitHub issues, send emails
- Why: Standardized tool definition and invocation
- Benefit: Safe, auditable tool execution

**5. RAG (Retrieval-Augmented Generation)**
- Example: Chatbot searching company docs before answering
- Why: MCP handles retrieval, LLM handles generation
- Benefit: Clean separation of concerns

**6. Agentic AI Systems**
- Example: AI agent researching and writing reports
- Why: Agent needs multiple tools and data sources
- Benefit: Orchestrate complex multi-step workflows

**7. Enterprise AI Applications**
- Example: AI assistant for internal employees
- Why: Secure access to proprietary data
- Benefit: Controlled, auditable data access

**8. Development Tools**
- Example: AI code assistant (like GitHub Copilot)
- Why: Access project files, git history, docs
- Benefit: Context-aware code suggestions

### âŒ **Don't Use MCP When:**

**1. Simple, One-Off Scripts**
- Problem: Just need to call OpenAI API once
- Better: Direct API call
- Why: MCP adds unnecessary complexity

**2. Static Data Only**
- Problem: Data never changes, can be embedded in prompt
- Better: Include data directly in prompt
- Why: No need for dynamic context fetching

**3. No Tool/Resource Needs**
- Problem: AI just generates text, no external data needed
- Better: Standard LLM API
- Why: MCP is for context integration

**4. Real-Time Streaming Critical**
- Problem: Sub-millisecond latency required
- Better: Direct in-memory access
- Why: Protocol overhead adds latency

**5. Legacy Systems (Can't Use JSON-RPC)**
- Problem: System only supports SOAP/XML
- Better: Custom adapter or different protocol
- Why: MCP requires JSON-RPC 2.0

## ğŸ“Š How It Works

**MCP Communication Flow:**

```
1. Initialization
   Client â†’ Server: initialize (capabilities)
   Server â†’ Client: initialized (capabilities)

2. Discovery
   Client â†’ Server: resources/list
   Server â†’ Client: [list of resources]
   
   Client â†’ Server: tools/list
   Server â†’ Client: [list of tools]

3. Usage
   Client â†’ Server: resources/read (uri)
   Server â†’ Client: {resource contents}
   
   Client â†’ Server: tools/call (name, arguments)
   Server â†’ Client: {tool result}

4. Shutdown
   Client â†’ Server: close connection
```

**Message Format (JSON-RPC 2.0):**
```json

}

// Response
{
  "jsonrpc": "2.0",
  "id": 1,
  "result": {
    "contents": [
      {
        "uri": "file:///data/customers.csv",
        "mimeType": "text/csv",
        "text": "name,email,status\nJohn,john@example.com,active"
      }
    ]
  }
}
```

````python
# MODEL CONTEXT PROTOCOL (MCP) - COMPLETE EXAMPLE

print("="*80)
print("MODEL CONTEXT PROTOCOL (MCP) - COMPREHENSIVE GUIDE")
print("="*80)

import json
from typing import Dict, List, Any, Optional
from dataclasses import dataclass, asdict
from enum import Enum

# 1. MCP MESSAGE STRUCTURE
print("\n1. MCP MESSAGE STRUCTURE (JSON-RPC 2.0)")
print("-"*80)

@dataclass
class MCPRequest:
    """MCP Request Message (JSON-RPC 2.0)"""
    jsonrpc: str = "2.0"
    id: int = 1
    method: str = ""
    params: Optional[Dict[str, Any]] = None

@dataclass
class MCPResponse:
    """MCP Response Message (JSON-RPC 2.0)"""
    jsonrpc: str = "2.0"
    id: int = 1
    result: Optional[Dict[str, Any]] = None
    error: Optional[Dict[str, Any]] = None

# Example: List resources request
list_resources_request = MCPRequest(
    id=1,
    method="resources/list",
    params={}
)

print("Example Request (resources/list):")
print(json.dumps(asdict(list_resources_request), indent=2))

# Example: Response with resources
list_resources_response = MCPResponse(
    id=1,
    result={
        "resources": [
            {
                "uri": "file:///data/customers.csv",
                "name": "Customer Database",
                "mimeType": "text/csv",
                "description": "All customer records"
            },
            {
                "uri": "file:///docs/manual.pdf",
                "name": "Product Manual",
                "mimeType": "application/pdf"
            }
        ]
    }
)

print("\nExample Response:")
print(json.dumps(asdict(list_resources_response), indent=2))

# 2. MCP SERVER SIMULATION
print("\n2. SIMPLE MCP SERVER SIMULATION")
print("-"*80)

class SimpleMCPServer:
    """Simple MCP Server implementation for demonstration"""
    
    def __init__(self, name: str):
        self.name = name
        self.resources = {}
        self.tools = {}
        self.request_id_counter = 0
        
    def add_resource(self, uri: str, name: str, content: str, mime_type: str = "text/plain"):
        """Add a resource to the server"""
        self.resources[uri] = {
            "uri": uri,
            "name": name,
            "content": content,
            "mimeType": mime_type
        }
        
    def add_tool(self, name: str, description: str, handler):
        """Add a tool to the server"""
        self.tools[name] = {
            "name": name,
            "description": description,
            "handler": handler,
            "inputSchema": {
                "type": "object",
                "properties": {},
                "required": []
            }
        }
    
    def handle_request(self, request: Dict[str, Any]) -> Dict[str, Any]:
        """Handle incoming MCP request"""
        method = request.get("method")
        params = request.get("params", {})
        request_id = request.get("id", 0)
        
        # Route to appropriate handler
        if method == "initialize":
            return self._handle_initialize(request_id)
        elif method == "resources/list":
            return self._handle_list_resources(request_id)
        elif method == "resources/read":
            return self._handle_read_resource(request_id, params)
        elif method == "tools/list":
            return self._handle_list_tools(request_id)
        elif method == "tools/call":
            return self._handle_call_tool(request_id, params)
        else:
            return self._error_response(request_id, -32601, "Method not found")
    
    def _handle_initialize(self, request_id: int) -> Dict[str, Any]:
        """Handle initialize request"""
        return {
            "jsonrpc": "2.0",
            "id": request_id,
            "result": {
                "protocolVersion": "0.1.0",
                "serverInfo": {
                    "name": self.name,
                    "version": "1.0.0"
                },
                "capabilities": {
                    "resources": {},
                    "tools": {}
                }
            }
        }
    
    def _handle_list_resources(self, request_id: int) -> Dict[str, Any]:
        """Handle resources/list request"""
        resources_list = [
            {k: v for k, v in resource.items() if k != "content"}
            for resource in self.resources.values()
        ]
        
        return {
            "jsonrpc": "2.0",
            "id": request_id,
            "result": {
                "resources": resources_list
            }
        }
    
    def _handle_read_resource(self, request_id: int, params: Dict[str, Any]) -> Dict[str, Any]:
        """Handle resources/read request"""
        uri = params.get("uri")
        
        if uri not in self.resources:
            return self._error_response(request_id, -32002, f"Resource not found: {uri}")
        
        resource = self.resources[uri]
        
        return {
            "jsonrpc": "2.0",
            "id": request_id,
            "result": {
                "contents": [
                    {
                        "uri": resource["uri"],
                        "mimeType": resource["mimeType"],
                        "text": resource["content"]
                    }
                ]
            }
        }
    
    def _handle_list_tools(self, request_id: int) -> Dict[str, Any]:
        """Handle tools/list request"""
        tools_list = [
            {
                "name": tool["name"],
                "description": tool["description"],
                "inputSchema": tool["inputSchema"]
            }
            for tool in self.tools.values()
        ]
        
        return {
            "jsonrpc": "2.0",
            "id": request_id,
            "result": {
                "tools": tools_list
            }
        }
    
    def _handle_call_tool(self, request_id: int, params: Dict[str, Any]) -> Dict[str, Any]:
        """Handle tools/call request"""
        tool_name = params.get("name")
        arguments = params.get("arguments", {})
        
        if tool_name not in self.tools:
            return self._error_response(request_id, -32002, f"Tool not found: {tool_name}")
        
        tool = self.tools[tool_name]
        
        try:
            result = tool["handler"](**arguments)
            
            return {
                "jsonrpc": "2.0",
                "id": request_id,
                "result": {
                    "content": [
                        {
                            "type": "text",
                            "text": str(result)
                        }
                    ]
                }
            }
        except Exception as e:
            return self._error_response(request_id, -32000, f"Tool execution error: {str(e)}")
    
    def _error_response(self, request_id: int, code: int, message: str) -> Dict[str, Any]:
        """Create error response"""
        return {
            "jsonrpc": "2.0",
            "id": request_id,
            "error": {
                "code": code,
                "message": message
            }
        }

# Create and configure server
server = SimpleMCPServer("Demo MCP Server")

# Add resources
server.add_resource(
    uri="file:///data/customers.csv",
    name="Customer Database",
    content="name,email,status\nJohn Doe,john@example.com,active\nJane Smith,jane@example.com,inactive",
    mime_type="text/csv"
)

server.add_resource(
    uri="file:///docs/readme.md",
    name="README",
    content="# My Project\n\nThis is a demo MCP server.",
    mime_type="text/markdown"
)

# Add tools
def search_customers(query: str) -> str:
    """Search for customers by name or email"""
    customers = [
        {"name": "John Doe", "email": "john@example.com"},
        {"name": "Jane Smith", "email": "jane@example.com"}
    ]
    results = [c for c in customers if query.lower() in c["name"].lower() or query.lower() in c["email"].lower()]
    return json.dumps(results, indent=2)

def calculate(expression: str) -> str:
    """Calculate mathematical expression"""
    try:
        result = eval(expression, {"__builtins__": {}}, {})
        return f"Result: {result}"
    except Exception as e:
        return f"Error: {str(e)}"

server.add_tool(
    name="search_customers",
    description="Search for customers by name or email",
    handler=search_customers
)

server.add_tool(
    name="calculate",
    description="Calculate a mathematical expression",
    handler=calculate
)

print("MCP Server initialized with:")
print(f"  - {len(server.resources)} resources")
print(f"  - {len(server.tools)} tools")

# 3. TEST MCP INTERACTIONS
print("\n3. TESTING MCP CLIENT-SERVER INTERACTIONS")
print("-"*80)

# Test 1: Initialize
print("\nTest 1: Initialize Connection")
init_request = {
    "jsonrpc": "2.0",
    "id": 1,
    "method": "initialize",
    "params": {
        "protocolVersion": "0.1.0",
        "clientInfo": {
            "name": "Demo Client",
            "version": "1.0.0"
        }
    }
}

print("Request:")
print(json.dumps(init_request, indent=2))

init_response = server.handle_request(init_request)
print("\nResponse:")
print(json.dumps(init_response, indent=2))

# Test 2: List Resources
print("\n\nTest 2: List Available Resources")
list_request = {
    "jsonrpc": "2.0",
    "id": 2,
    "method": "resources/list",
    "params": {}
}

print("Request:")
print(json.dumps(list_request, indent=2))

list_response = server.handle_request(list_request)
print("\nResponse:")
print(json.dumps(list_response, indent=2))

# Test 3: Read Resource
print("\n\nTest 3: Read Specific Resource")
read_request = {
    "jsonrpc": "2.0",
    "id": 3,
    "method": "resources/read",
    "params": {
        "uri": "file:///data/customers.csv"
    }
}

print("Request:")
print(json.dumps(read_request, indent=2))

read_response = server.handle_request(read_request)
print("\nResponse:")
print(json.dumps(read_response, indent=2))

# Test 4: List Tools
print("\n\nTest 4: List Available Tools")
tools_request = {
    "jsonrpc": "2.0",
    "id": 4,
    "method": "tools/list",
    "params": {}
}

print("Request:")
print(json.dumps(tools_request, indent=2))

tools_response = server.handle_request(tools_request)
print("\nResponse:")
print(json.dumps(tools_response, indent=2))

# Test 5: Call Tool
print("\n\nTest 5: Call Tool (search_customers)")
call_request = {
    "jsonrpc": "2.0",
    "id": 5,
    "method": "tools/call",
    "params": {
        "name": "search_customers",
        "arguments": {
            "query": "john"
        }
    }
}

print("Request:")
print(json.dumps(call_request, indent=2))

call_response = server.handle_request(call_request)
print("\nResponse:")
print(json.dumps(call_response, indent=2))

# Test 6: Error Handling
print("\n\nTest 6: Error Handling (Invalid Resource)")
error_request = {
    "jsonrpc": "2.0",
    "id": 6,
    "method": "resources/read",
    "params": {
        "uri": "file:///nonexistent.txt"
    }
}

print("Request:")
print(json.dumps(error_request, indent=2))

error_response = server.handle_request(error_request)
print("\nResponse (Error):")
print(json.dumps(error_response, indent=2))

print("\n" + "="*80)
print("SUMMARY")
print("="*80)
print("âœ“ MCP standardizes AI-to-context-provider communication")
print("âœ“ Three core primitives: Resources, Tools, Prompts")
print("âœ“ JSON-RPC 2.0 protocol for all messages")
print("âœ“ Client-server architecture (AI app = client, data = server)")
print("âœ“ Resources: Read-only data sources (files, databases, APIs)")
print("âœ“ Tools: Actions the LLM can invoke (search, create, update)")
print("âœ“ Transport: stdio (local), HTTP/SSE (remote), WebSocket")
print("âœ“ Use for: AI assistants, RAG, agentic AI, tool use")
print("âœ“ Security: Sandboxed, controlled access to resources")
print("âœ“ Open protocol: Growing ecosystem of servers and clients")
print("="*80)
````

**Component Roles:**

**1. MCP Client**
- Initiates connections to servers
- Discovers available resources and tools
- Sends requests, receives responses
- Example: Claude Desktop app

**2. MCP Server**
- Exposes resources, tools, and prompts
- Processes client requests
- Manages access to underlying data/services
- Example: Filesystem server, GitHub server

**3. Protocol Layer**
- JSON-RPC 2.0 message format
- Request ID correlation
- Error codes and handling

**4. Transport Layer**
- stdio: Local processes, simple IPC
- HTTP/SSE: Remote servers, client-initiated
- WebSocket: Bidirectional, real-time

## ğŸ¯ Why This Architecture?

### **Design Principles:**
1. **Separation of Concerns** - Clear boundaries between layers
2. **Transport Agnostic** - Protocol works over any transport
3. **Extensibility** - Easy to add new resource types, tools
4. **Security by Design** - Server controls access, not client
5. **Simplicity** - JSON-RPC 2.0 is well-understood
6. **Interoperability** - Any client works with any server

### **Advantages:**
1. **Modular** - Swap transport layers without changing logic
2. **Scalable** - Add servers without modifying client
3. **Testable** - Mock servers for testing
4. **Language-Agnostic** - JSON works in all languages
5. **Debuggable** - Human-readable messages
